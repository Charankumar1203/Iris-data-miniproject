# preparing our data for modeling 
splitting the data back out into plain NumPy arrays

```python
x_train = df_train.drop(columns=["target", "target_name"]).values
y_train = df_train["target"].values
```

# modeling - simple manual model
manually look at our data and decide some coutoff points for classification

```python
def single_feature_prediction(petal_length):
    if petal_length < 2.5:
        return 0
    elif petal_length < 4.8:
        return 1
    else:
        return 2
```

```python
df_train.columns
```

```python
x_train[:, 2]
```

```python
manual_y_predictions = np.array([single_feature_prediction(val) for val in x_train[:, 2]])
```

```python
manual_model_accuracy = np.mean(manual_y_predictions == y_train)
```

```python
print(f"Manual model accuracy: {manual_model_accuracy * 100:.2f}%")
```

# Modeling- logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

### Using a Validation set to evaluate our model

```python
model = LogisticRegression(max_iter=200)
```

```python
# xt = x_train and xv = x_validation
xt, xv, yt, yv = train_test_split(x_train, y_train, test_size=0.25)
```

```python
model.fit(xt, yt)
```

```python
y_pred = model.predict(xv)
```

```python
np.mean(y_pred == yv)
```

```python
model.score(x_train,y_train) # NEVER EVALUATE YOUR MODEL ON THE SAME DATA USED FOR TRAINING.
```

```python
model.score(xv,yv)
```

## Using Cross validation

```python
from sklearn.model_selection import cross_val_score, cross_val_predict
```

```python
model = LogisticRegression(max_iter=200)
```

```python
accuracies = cross_val_score(model, x_train, y_train, cv=5, scoring="accuracy")
```

```python
np.mean(accuracies)
```

```python
y_pred = cross_val_predict(model, x_train, y_train, cv=5)
```

```python
predicted_correctly_mask = y_pred == y_train
```

```python
not_predicted_correctly = ~predicted_correctly_mask
```

```python
x_train[not_predicted_correctly]
```

```python
df_predictions = df_train.copy()
```

```python
df_predictions["correct_predictions"] = predicted_correctly_mask
```

```python
df_predictions["prediction"] = y_pred
```

```python
df_predictions["predictions_label"] = df_predictions["prediction"].map({0: "setosa", 1: "versicolor", 2: "virginica"})
```

```python
df_predictions.head()
```

```python
sns.scatterplot(x="petal length (cm)", y="petal width (cm)", hue="predictions_label", data=df_predictions)
```

```python
sns.scatterplot(x="petal length (cm)", y="petal width (cm)", hue="target_name", data=df_predictions)
```

```python
def plot_incorrect_predictions(df_predictions, x_axis_feature, y_axis_feature):
    fig, axs=plt.subplots(2,2,figsize = (10,10))
    axs = axs.flatten()
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature, hue="predictions_label", data=df_predictions, ax=axs[0])
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature, hue="target_name", data=df_predictions, ax=axs[1])
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature, hue="correct_predictions", data=df_predictions, ax=axs[2])
    axs[3].set_visible(False)
    plt.show()
```

```python
plot_incorrect_predictions(df_predictions, "petal length (cm)", "petal width (cm)")
```
